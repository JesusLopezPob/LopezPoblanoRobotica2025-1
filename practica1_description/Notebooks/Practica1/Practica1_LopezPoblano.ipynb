## Objetivo

El objetivo de esta práctica es que el alumno comprenda, interprete y modifique la información básica de los formatos de Universal Robot Description File (URDF).

## Previo

- **Definición de URDF:**
En RosDocumentation, se le define al URDF como un formato para describir la geometria del robot que utiliza ROS, se le conoce por sus siglas (Unified Robot Description Format))[[1]](#1).
- **Tipos de juntas que están definidas dentro del formato URDF:**
Una junta es la conexion entre 2 eslabones, siendo que esta la fixed (no permite movimiento), prismatic(que se desliza a lo largo de un eje), revolute(que rota sobre un eje con un rango limitado de movimiento), continuous(que rota sobre un eje sin un limite en su rango de movimiento),planar(que se mueve en un plano ortogonal a un eje) y floating(puede moverse en los grados de libertad)[[2]](#2)[[4]](#4).
- **Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS:**
En el URDF se pueden ocupar formas primitivas o mesh para poder visualizar al robot, siendo las formas primitivas el box(se define las dimensiones de las longitudes para definir la caja, siendo el origen el centro del mismo), cylinder(aqui especificamos el radio y la altura, siendo el origen el centro de la misma) y sphere(se define unicamente el radio,siendo el centro de la esfera su origen)[[3]](#3) .
- **Que es un nodo y que es un topico en ROS 2:** Los nodos se encargan de una simple y modular tarea, es decir, puede controlar un motor, leer sensores, etc. [[5]](#5), ademas un nodo puede mandar o recibir de varios Topics, siendo los Topics una fomra de trasladar informacion entre nodos[[6]](#6) en otras palabras, el nodo es un programa que realiza una tarea en especifico y puede mandar y/o recibir datos de otros nodos mediante un topic de manera continua.
- **Que es RVIZ:** Rviz es un visualizador 3D para frameworks de ROS [[7]](#7).


## Desarrollo

Primero se creo el *workspace* donde se desarrollara la practica, para ello se ejecutaron los siguientes comandos en la consola:

- **cd ROS2Dev:**
- **mkdir -p /robot_2025_ws**
- **cd robot_2025_ws**
- **colcon build**


El primer comando nos permite acceder a la capeta ROS2Dev y de hay crear un directorio con el nombre robot_2025_ws y al mismo tiempo un directorio llamado src dentro, despues nos ubicamos en el directorio del workspace y con *colcon build* compila el workspace.

Despues nos dirigimos a la carpeta de source en nuestro **workspace** con *cd src/*. Ya en el directorio de src, creamos un paquete para ROS2 con este comando: **ros2 pkg create --build-type  ament_cmake practica1_description**; siendo ***ros2 pkg create*** el comando para crear los paquetes en ROS2, ***--build-type ament_cmake*** indica el tipo de sistema de construcción que utilizará el paquete, en nuestro caso el de **ament_cmake** el sistema de compilación estándar en ROS 2. Finalmente el ultimo parametro es el nombre de  nuestro paquete.

Ya creado el paquete, debemos crear tres carpetas *urdf*,*rviz* y *launch*, para ello ejecutamos el comando *mkdir urdf rviz launch* en el directorio del paquete.

Con el comando *code .*, nos desplegara el visual code con nuestro directorio del paquete, siendo que crearemos 2 archivos: uno la carpeta  launch  como *basic_link_display.launch.xml* y otro en la carpeta URDF como *bsic_linl.urdf*

Codigo del *basic_link_display.launch.xml* :
```xml
<launch>
    <let name="urdf_path"
        value="$(find-pkg-share practica1_description)/urdf/basic_link.urdf" />
    
    <let name="rviz_config"
        value="$(find-pkg-share practica1_description)/rviz/basic_link_display.rviz" />
    
    <node pkg="robot_state_publisher" exec="robot_state_publisher">
        <param name="robot_description"
            value="$(command 'xacro $(var urdf_path)')" />
    </node>
    <node pkg = "joint_state_publisher_gui" exec = "joint_state_publisher_gui" />
    
    <node pkg = "rviz2" exec = "rviz2" output = "screen"
        args = " -d $(var rviz_config)" />
        

</launch>
```

Codigo del *basic_link.urdf* hecho en clase:

```xml
<?xml version="1.0" encoding="utf-8"?>

<robot name="basic_link">

    <link name="base_link">
        <visual>
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>
            <material name="brown">
                <color rgba="0.1 0.43 0.43 1.0" />
            </material>      
        </visual>
        <collision>
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>          
        </collision>
        <inertial>
            <mass value="0.5"/>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

    <link name="link_1">
        <visual>
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <material name="red">
                <color rgba="0.5 0.0 0.0 1.0" />
            </material>      
        </visual>
        <collision>
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <inertial>
            <mass value="0.3"/>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

   
    <joint name="base_link_to_link_1" type="revolute">
        <parent link="base_link"/>
        <child link="link_1"/>
        <axis xyz="1 0 0"/>
        <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
        <origin xyz="0.05 0 0.15" rpy="0 0 0"/>
    </joint>


</robot>
```

Una vez terminado de programar, nos regresamos al directorio del workspace,descargamos 2 pluggins con *sudo apt install ros-humble-joint-state-publisher-gui* y *sudo apt install ros-humble-xacro*, para que estos pluggings hagan efecto hay que ejecutar *sudo apt-get update* para actualizar los paquetes. Despues compilamos el workspace con estos comandos:

- **colcon build**
- **source install/setup.bash**

Si se siguieron los pasos, no existirian errores para ejecutar el urdf en Rivz, para ello se ejecuta ***ros2 launch practica1_description basic_link_display.launch.xml***, esperamos unos segundos y se abrira el RViz, primero debemos configurar el "fixed frame", damos click en la lista y seleccionamos *base_link*. Despues debemos agregar nuestro urdf, para ello nos vamos al boton *add* y buscar "RobotModel", damos click y aparecera en *displays*, presionamos la flecha que esta alado de "RobotModel" y nos desplegara una lista de configuracion, buscamos ***DescriptionSource*** y seleccionamos *file* ya que estara la opcion por defecto *topic* y en ***descriptionFile*** le damos click en los tres puntos para seleccionar la ubicacion de nuestro urdf.
Finalmente podremos ver nuestro URDF.

Como parte de nuestra practica es agregar otro eslabon mas a nuestro robot, ya que hasta el momento. nuestro robot solamente cuenta con una junta y 2 eslabones. Para ello nos vamos al archivo que nos describe nuestro robot, es decir, el **basic_link.urdf**

En el archivo URDF, agregamos otro eslabon, nombrandolo como *link_2*, definiendo su masa, propiedades inerciales, visual,etc como los demas eslabones que desarrollamos en clase y agregamos otra junta , similar a la junta que programamos en clase.
Siendo el codigo agregado para el eslabon *Link_2* en **basic_link.urdf**:
```xml
 <link name="link_2">
        <visual>
            <origin xyz="0.025 0.0 0.15" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.03 0.03 0.3"/>             
            </geometry>
            <material name="yellow">
                <color rgba="0.5 0.5 0.0 1.0" />
            </material>      
        </visual>
        <collision>
            <origin xyz="0.0 -0.20 0.0" rpy="0.0 0.0 0.0"/>
            <geometry>
                <box size="0.03 0.03 0.3"/>             
            </geometry>          
        </collision>
        <inertial>
            <mass value="0.1"/>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>
```
Y el codigo agregado para agregar otra junta en **basic_link.urdf**:
```xml
     <joint name="link_1_to_link_2" type="revolute">
        <parent link="link_1"/>
        <child link="link_2"/>
        <axis xyz="1 0 0"/>
        <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="100.0"/>
        <origin xyz="-0.04 0.0 0.48" rpy="0 0 0"/>
    </joint>
```

Adicionalmente podemos ver un diagrama de como interactuan los nodos para la visualizacion del urdf.

![Diagrama](Images/Diagrama.png) 

El diagrama  muestra la interacción entre los nodos del sistema basado en ROS 2, que gestiona la publicación de transformaciones y el estado del robot. El nodo */robot_state_publisher* recibe la información sobre las juntas del robot del tópico */joint_states* y la descripción completa del robot de */robot_description*. Calcula las transformaciones espaciales a partir de estos datos y publica los resultados en los tópicos */tf* y */tf_static*.

Por un lado, el nodo /joint_state_publisher publica los estados de las juntas del robot, proporcionando toda la información necesaria para que el nodo *robot_state_publisher* cree las transformaciones correctas.

Finalmente, */transform_listener_impl* se suscribe a los tópicos */tf* y */tf_static* para recibir y procesar las transformaciones, lo que permite integrar esta información con aplicaciones como RViz para la visualización en tiempo real del robot. Por lo tanto, este flujo es fundamental para obtener una representación adecuada y actualizada del estado del robot, necesaria para tareas de simulación y control.

## Conclusiones
Durante esta práctica, se logró obtener información sobre la estructura y funcionalidad de los archivos Universal Robot Description File (URDF), que son fundamentales para describir la estructura de un robot en el contexto de ROS 2. Aprendí a interpretar los componentes principales del URDF, como los enlaces (*link*) y las articulaciones (*joint*), comprendiendo las relaciones y transformaciones en el modelo del robot. Esto me proporcionó una base sólida para entender cómo estas descripciones influyen en la representación y funcionamiento del robot en un entorno simulado.

Ahora soy capaz de editar archivos URDF para personalizar modelos robóticos.
